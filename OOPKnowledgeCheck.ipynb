```python
# What is the purpose of the __str__ method in a Python class?

# To Define a custom string respresentation of an object

# Explanation - The __str__ method in a Python class is used to define a custom string representation of an object. 
# This method is called by the built-in str() function and the print function, providing a readable and informative 
# string that represents the object.
```

```python
# In the context of Object-Oriented Programming (OOP) in Python, what is the purpose of the self parameter in class methods?

# 

# The self parameter in class methods is used to refer to the instance of the class on which the method is being called. 
# It allows access to the instance variables and other methods of the class. 
# Without self, it is not possible to access or modify the attributes of the specific object.
```

```python
# What is the purpose of the __init__ method in a Python class?

# The __init__ method in a Python class is a special method called a constructor. It is automatically called when a new instance of the class is created, 
# and it is used to initialize the attributes of the new object with the values provided as arguments.
```
