```python

# Continuously read strings from a user until an empty string is entered.

# Store all strings in a list.

# Sort the the list in alphabetical order and then print it.

## Example input:
#I
#really
#love
#python

# You should print:

# ['I', 'love', 'python', 'really']

strings = []
while True:
    user_input = input()
    if user_input == "":
        break
    strings.append(user_input)
strings.sort()
print(strings) 
```
