```python
# Complete the function q9 which takes one parameter: an integer (miltime) representing military time and 
# returns a string containing the greeting of the day.

  #  0300-1159 "Good Morning"
   # 1200-1559 "Good Afternoon"
    #1600-2059 "Good Evening"
    # 2100-0259 "Good Night"

# NOTE: the actual time 0305 would be entered as 305 into the function

# Example:

# q9(305)

# Returns: Good Morning


def q9(miltime):
    if miltime >= 300 and miltime <= 1159:
        return ("Good Morning")
    
    elif miltime >= 1200 and miltime <= 1559:
        return ("Good Afternoon")
    
    elif miltime >= 1600 and miltime <= 2059:
        return ("Good Evening")

    else:
        return ("Good Night")


```

```python
# Complete the function q5 which takes two parameters: integer and limit, and print a list of multiples of the 
# integer up to and including the limit.

# For example, if integer = 3 and limit = 30, the printed list 
# should be [0, 3, 6, 9, 12, 15, 18, 21, 24, 27, 30]. Note, 0 is a multiple of any integer except 0 itself.

def q5(integer, limit):
    multiples = []
    for i in range(0, limit + 1, integer):
        multiples.append(i)
    return list(multiples)

q5(3,30)
```

```python
#Define a function called findBigger that takes two parameters, both of which should be integers.  
#The function should compare the two numbers and return First if the first number is bigger, Second if the second number is bigger, 
# and Equal if the numbers are equal.

Sample output:
findBigger(3,4) --> Second


def findBigger (First, Second):
    if First > Second:
        return "First"
    elif First < Second:
        return "Second"
    else:
        return "Equal"
```
