```python
# Define the function FreeRange(start, stop). 
# The function returns a range specified by the parameters with a step of +1 if start < stop. 
# Otherwise the function returns a range with a step of -1 if start > stop

#Define the function FreeRange(start, stop). The function returns a range specified by the parameters with a step of +1 if start < stop. 
#Otherwise the function returns a range with a step of -1 if start > stop

def FreeRange(start, stop):
    if start < stop:
        return range(start, stop)
    elif start > stop:
        return range(start, stop, -1)
```

```python
# Use python to complete the function below that will perform the following:

#     Given a mixed case string as parameter s
#     Capitalize every letter with an even index within the string.
#     Lowercase every letter with an odd index within the string.
#     Return the resulting string.
#     Example:
#         leetString("ABCDEF ghijkl")
#         Return: "AbCdEf gHiJkL"


def leetString(s):
    pass
    result_chars = []
    for i, char in enumerate(s):
        if i % 2 == 0:
            result_chars.append(char.upper())
        else:
            result_chars.append(char.lower())
    return "".join(result_chars)
```

```python
# Define the function MatrixProduct(matrix, multiplier) where matrix is a nested list and multiplier is a number.

# The function should multiply each individual item found inside matrix by the multiplier and then return the updated matrix

# Example:

# MatrixProduct([[1,2,3],[4,5,6],[7,8,9]],2)

# Returns: 

# [[2, 4, 6], [8, 10, 12], [14, 16, 18]]

def MatrixProduct(matrix, multiplier):
    for lst in range(len(matrix)):
        emplst = []
        for value in matrix[lst]:
            value *= multiplier
            emplst.append(value)
        matrix[lst] = emplst
    return matrix

```

```python
# Write a function named name_game that takes one parameter fname, a file of names, separated by line breaks, 
# and return a dictionary where the key is the first letter of the name 
# nd the value is the first instance of a name that begins with that first letter.

def name_game(fname):
    name_dict = {}
    try:
        with open(fname, 'r') as file:
            for line in file:
                name = line.strip()  # Remove leading/trailing whitespace and newline characters
                if name:  # Ensure the line is not empty
                    first_letter = name[0].upper()  # Get the first letter and convert to uppercase
                    if first_letter not in name_dict:
                        name_dict[first_letter] = name
    except FileNotFoundError:
        print(f"Error: The file '{fname}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
    return name_dict
```

```python
#####################Dictionaries and Sets - End of Lecture CTF Challenges ##############################

# Complete the function below to perform the following:

#     fname is a file containing words that are separated by at least one character of whitespace.
#     A list of words is provided as the parameter words.
#     Return a list of all the words found in the file that are NOT contained in the list of words in parameter.

# You can use act24.txt as a test file to complete this challenge.

# Example: 

# diffwords('act24.txt', ['this', 'is', 'a', 'list', 'of', 'words'])

# Returns:  ['in', 'file', 'the']



def diffwords(fname, words):
    unique_words_in_file = []
    with open(fname, 'r') as file:
            file_content = file.read()
            file_words = file_content.split()  # Splits by any whitespace
            
            # Convert the input 'words' list to a set for efficient lookup
            words_set = set(words)

            for word in file_words:
                if word not in words_set:
                    unique_words_in_file.append(word)
                    
    return unique_words_in_file
```

```python
# Count the occurrences of each word in the file: act25.txt.

# Create a dictionary that contains each word in the file as a key and the value for each key will contain the number of times each words is found in the file. Words will be separated by one or more whitespace characters spread over multiple lines.

# Return the dictionary.

# Example:

# the file contains the following sentence:  a dog jumped a fence

# the returned dictionary would contain:

# {'a':2, 'dog':1, 'jumped':1, 'fence':1}

def count_words(filepath):
    word_counts = {}
    with open(filepath, 'r') as file:
        for line in file:
            # Split the line into words, handling multiple whitespace characters
            words = line.split()
            for word in words:
                # Remove punctuation if necessary (optional, but good practice for word counting)
                word = word.strip()
                if word:  # Ensure the word is not empty after stripping
                    word_counts[word] = word_counts.get(word, 0) + 1
    return word_counts

```

```python
# Read all lines from file in filepath and return a list of all lines in case-insensitive ASCII order. Remove all linebreaks before sorting.

# You can use the following file for testing purposes: file1.txt

def sort_ascii(filepath):
    lines = []
    try:
        with open(filepath, 'r') as file:
            for line in file:
                lines.append(line.strip())  # Remove leading/trailing whitespace, including linebreaks
    except FileNotFoundError:
        print(f"Error: File not found at {filepath}")
        return []

    # Sort the list case-insensitively
    lines.sort(key=str.lower)
    return lines
```

```python
# Read all lines from file in filepath and return a list of all lines sorted longest to shortest. Remove all linebreaks before sorting.

# You can use the following file for testing: file1.txt

def sort_length(filepath):
    with open(filepath, 'r') as file:
        lines = file.readlines()
        counter = 0
        for line in lines:
            lines[counter] = line.strip('\n')
            counter += 1
    lines.sort(key=len, reverse=True)
    return lines
```

```python
# Read all lines from file in filepath and return a list of all lines sorted numerically by the number at character positions 9 to 14 in each line. (Hint: Note the difference between character positions and index positions)

# Remove all line breaks before sorting.

# Example: The embedded number is 561234 below.

# Here is a561234 long line of text from the file.

# You can use act29.txtfor testing.

import re

def sort_embedded(filepath):
    """
    Reads lines from a file, removes line breaks, and sorts them numerically
    by the 6-digit number embedded after a letter like 'a'.

    Args:
        filepath (str): The path to the file.

    Returns:
        list: A list of the sorted lines.
    """
    try:
        with open(filepath, 'r') as file:
            lines = [line.strip() for line in file.readlines()]

        def get_sort_key(line):
            match = re.search(r'a(\d{6})', line)
            if match:
                return int(match.group(1))
            else:
                return float('inf') # Return a high value for lines without the pattern

        sorted_lines = sorted(lines, key=get_sort_key)
        return sorted_lines
    except FileNotFoundError:
        print(f"Error: The file at {filepath} was not found.")
        return []
```

```python
# Create a class named Chest that can store a large number of items in to a list for an explorer. When initializing a Chest instance, it can store a variable number of items. 
# The chest has a function called add that can store an additional random number of items. 
# It also has a function called drop that can remove a variable number of items from the chest. The empty function removes all the items inside the chest. 
# When you print the chest instance, it will return Chest: [itemX, itemY, itemZ].

# For example, by calling chest1 = Chest('swords', 'bows', 'daggers') and then print(chest1), it will return Chest: ['swords', 'bows', 'daggers'].

class Chest:

    def __init__(self, *items):

        self.items = list(items)

    def __repr__(self):
        return f"Chest: {self.items}"

    def add(self, *additional_items):
        self.items.extend(additional_items)

    def drop(self, *items_to_remove):
        for item in items_to_remove:
            if item in self.items:
                self.items.remove(item)

    def empty(self):
        self.items.clear()
```

```python
# Prime numbers are important in cryptography.  A number is prime if it is only divisible by itself and 1. 
# Define a function called isPrime that takes one parameter (an integer) and returns True if the parameter is prime and False otherwise.

def isPrime(number):

    if number <= 1:
        return False  
    if number == 2:
        return True  
    if number % 2 == 0:
        return False  

    i = 3
    while i * i <= number:
        if number % i == 0:
            return False 
        i += 2  

    return True
```

```python
# Define a function called listSquares that takes one parameter which is a string of space separated numbers. 
# The function should process each number and if the number is non-negative, square the number and add it to a list.  
# If a negative number is encountered the function should stop processing numbers and return the list 
# (the square of the negative number should not be included in the list). If the first number is negative, then the function should return an empty list.

# Example output:
# listSquares("3 6 9 10 -5 4 9 1") should return [9, 36, 81, 100]
# listSquares("-13 5 17 11 14 16 18 7") should return []
def listSquares(numbers_string):
    squared_numbers = []
    numbers = numbers_string.split()
    try:
        if int(numbers[0]) < 0:
            return []
    except ValueError:
        return []

    for num_str in numbers:
        try:
            num = int(num_str)
            if num >= 0:
                squared_numbers.append(num * num)
            else:
                break
        except ValueError:
            break
    return squared_numbers

```

```python
# Define a function called cleanString which takes two parameters.  The first parameter is a string with multiple characters. 
# The second parameter is a one-character string. The function should remove all instances of the second parameter (case-insensitive) from the first parameter and 
# return this new string.

# Example output:
# cleanString('This Is A Sentence', 'i') should return "Ths s A Sentence"
# cleanString('This Is A Sentence', 'I') should return "Ths s A Sentence"

def cleanString(main_string, char_to_remove):
    lower_char_to_remove = char_to_remove.lower()
    cleaned_chars = []

    for char in main_string:
        if char.lower() != lower_char_to_remove:
            cleaned_chars.append(char)

    return "".join(cleaned_chars)
```

```python
# Define a class called BankAccount.  The class should have a method that initializes an instance of BankAccount. 
# This method should have parameters which provide values for the attributes accountNum (string), holderFirstName (string), holderLastName (string), 
# and balance (float).  This method should also set an attribute status (bool) to  True.  While accountNum is a string, 
# all characters should be digits.  The provided value for balance should be non-negative.  
# If the provided values for accountNum and balance are not correct then display the message "Account not created" and do not assign any values 
# to the attributes.

# Define a method called makeDeposit that takes one parameter (a non-negative float) representing the amount of money to be deposited in the account.
# If the account is closed (i.e. status is set to False) then display the message "Account is closed" and take no further action.  
# If the parameter value is not valid then display the message "Deposit value invalid" and take no further action.  
# Otherwise, update the balance attribute by adding the parameter value.

# Define a method called makeWithdrawal that takes one parameter (a non-negative float) representing the amount of money to withdraw from the account.
# If the account is closed (i.e. status is set to False) then display the message "Account is closed" and take no further action.  
# If the parameter value is not valid then display the message "Withdrawal value invalid" and take no further action.  
# If the parameter value is valid, but is greater that then balance value, then display the message "Insufficient funds" and take no further action. 
# Otherwise, update the balance attribute by subtracting the parameter value.

# Define a method called returnBalance.  If the account is closed, it should display the message "Account closed" and take no further action. 
# Otherwise, return the account balance formatted correctly (with a dollar sign and two decimal places).

# Define a method called closeAccount.  If the account is closed, it should display the message "Account is already closed".  
# Otherwise, set the status attribute to False.

# Define a special method for displaying an object in the form:
# accountNum
# holderFirstName holderLastName
# balance
# status


class BankAccount:
    def __init__(self, accountNum, holderFirstName, holderLastName, balance):
        try:
            # Attempt to convert balance to a float and validate it
            if accountNum.isdigit() and balance >= 0:
                self.accountNum = accountNum
                self.holderFirstName = holderFirstName
                self.holderLastName = holderLastName
                self.balance = balance
                self.status = True
            else:
                print("Account not created")
                self._clear_attributes()
        except (ValueError, TypeError):
            # The float conversion will raise a ValueError if it cannot parse the string.
            print("Account not created")
            self._clear_attributes()

    def _clear_attributes(self):
        self.accountNum = None
        self.holderFirstName = None
        self.holderLastName = None
        self.balance = None
        self.status = None
      
    def makeDeposit(self, amount):
        if not self.status:
            print("Account is closed")
            return
        if not isinstance(amount, (int, float)) or amount < 0:
            print("Deposit value invalid")
            return
        self.balance += amount

    def makeWithdrawal(self, amount):
        if not self.status:
            print("Account is closed")
            return
        if not isinstance(amount, (int, float)) or amount < 0:
            print("Withdrawal value invalid")
            return
        if amount > self.balance:
            print("Insufficient funds")
            return
        self.balance -= amount

    def returnBalance(self):
        if not self.status:
            print("Account is closed")
            return
        return f"${self.balance:.2f}"

    def closeAccount(self):
        if not self.status:
            print("Account is already closed")
            return
        self.status = False

    def __str__(self):
        # This is a special method used by print() and str()
        if self.accountNum is None:
            return "Invalid Account"
        return f"{self.accountNum}\n{self.holderFirstName} {self.holderLastName}\n${self.balance:.2f}\n{self.status}"

```

```python
# Define a function called charFreq that takes one parameter which is a filename.  Open the file and read the contents.  
# Determine how many times each alphabetic character (case insensitive) appears in the file.  Display the alphabetic 
# character that appears the most and the alphabetic character that appears the least with the number of times each appears in parentheses.

def charFreq(filename):
    char_counts = {}
    with open(filename, 'r') as file:
        content = file.read()
        for char in content:
            if 'a' <= char.lower() <= 'z':
                char_counts[char.lower()] = char_counts.get(char.lower(), 0) + 1

    most_frequent_char = max(char_counts, key=char_counts.get)
    least_frequent_char = min(char_counts, key=char_counts.get)

    print(f"Most: {most_frequent_char} ({char_counts[most_frequent_char]})")
    print(f"Least: {least_frequent_char} ({char_counts[least_frequent_char]})")
```

```python

# Complete the function q6 which takes two parameters: a dictionary (catalog) whose keys are product names and values are product prices per unit and 
# a list of tuples (order) of product names and quantities, computes and returns the total value of the order.

# Using this catalog:

# catalog = {'AMD Ryzen 5 5600X': 289.99,'Intel Core i9-9900K': 363.50,'AMD Ryzen 9 5900X': 569.99}

# And this order:

# order = [('AMD Ryzen 5 5600X', 5), ('Intel Core i9-9900K', 3)]

# Example:

# q6(catalog, order)

# Returns: 2540.45

# How the above result was computed:

# (289.99 * 5) + (363.50 * 3)


def q6(catalog, order):
    total_value = 0.0
    for product_name, quantity in order:
        if product_name in catalog:
            total_value += catalog[product_name] * quantity
    return total_value
```

```python
# Define a function called myRange that takes a variable number of arguments, but no more than 3.  
# The function should return a list of integers that approximates the range function.  If the number of arguments is 1, 
# then the function should return a list of integers starting at 0 and increments by 1 up to one less than the argument. If the number of arguments is 2, 
# then the function should return a list of integers starting at the first argument and increments by 1 up to one less than the second argument.  
# If the number of arguments is 3, then the function should return a list of integers starting at the first argument and increments by the third argument going 
# up to one less than the second argument.  If there are more than 4 arguments the function should return Too many arguments.

# Sample output:
# myRange(4) --> [0, 1, 2, 3]

def myRange(*args):
    num_args = len(args)

    if num_args > 3:
        return "Too many arguments"
    elif num_args == 1:
        stop = args[0]
        start = 0
        step = 1
    elif num_args == 2:
        start = args[0]
        stop = args[1]
        step = 1
    elif num_args == 3:
        start = args[0]
        stop = args[1]
        step = args[2]
    else:  # This case handles 0 arguments, though the problem implies at least 1.
        return [] # Or raise an error, depending on desired behavior for 0 arguments

    result = []
    current = start
    if step > 0:
        while current < stop:
            result.append(current)
            current += step
    elif step < 0:
        while current > stop:
            result.append(current)
            current += step
    return result
```

```python
# Define a function called milPhone that takes one parameter (a string).  The function should check that the parameter only contains letters (i.e. one word).  
# If the parameter does not contain only letters, then return the string Argument must only be letters. 
# If the parameter is in the correct form, then return a string where each letter has been replaced with its military phonetic equivalent and there is a space between each.  
# The returned string should be lowercase.

# Sample output:
# milPhone('hello hello') --> Argument must only be letters
# milPhone('hello') --> hotel echo lima lima oscar

milDict = {'a': 'alpha', 'b':'bravo', 'c':'charlie', 'd':'delta', 'e':'echo', 'f':'foxtrot', 'g':'golf', 'h':'hotel', 'i':'india', 'j':'juliet', 'k':'kilo', 'l':'lima', 'm':'mike', 'n':'november', 'o':'oscar', 'p':'papa', 'q':'quebec', 'r':'romeo', 's':'sierra', 't':'tango', 'u':'uniform', 'v':'victor', 'w':'whiskey', 'x':'xray', 'z':'zulu'}

def milPhone(word_string):
    if not word_string.isalpha():
        return "Argument must only be letters"
    result = []
    for char in word_string.lower():
        result.append(milDict[char])

    return ' '.join(result)
```

```python
# Define a function called uniqueChar that takes one parameter (a string).  The function should return a list of non-case sensitive characters that occur in the parameter, 
# however each character should appear only once in the list.  The list should be sorted in ascending order and only contain lowercase characters.

# Sample output:
# uniqueChar('helLo') --> ['e', 'h', 'l', 'o']

def uniqueChar(input_string):
    unique_chars = set()
    for char in input_string:
        if char.isalpha():  # Only consider alphabetic characters
            unique_chars.add(char.lower())  # Add lowercase version to set for uniqueness
    
    return sorted(list(unique_chars))
```

```python
# Define a function called compareVals that takes a variable number of parameters (which should be integers) and one keyword parameter called com whose default value is 0.  
# Then check each value in the variable number of parameters and if the value is not an integer, then the function should return the string Incorrect value.  
# Otherwise compare each value in the variable number of parameters to the the keyword parameter value.  
# If the value is strictly greater than the keyword value then add it to a list. Return the list of strictly greater values.

# Sample output:
# compareVals(1,2,3.1,4,com=2) --> Incorrect value
# compareVals(1,2,3,4,com=2) --> [3, 4]


def compareVals(*args, com=0):
    greater_values = []
    for val in args:
        if not isinstance(val, int):
            return "Incorrect value"
        if val > com:
            greater_values.append(val)
    return greater_values
```

```python
# Write a program that accepts input from the user.  The input will be consist of an integer, an operator, and an integer where each is separated by a space 
# (e.g. 3 + 3, 4 / 4, 5 * 2).  The acceptable operators are + (addition), - (subtraction), / (division), and * (multiplication). 
# If the input is not in the correct format then output Incorrect Format (includes neither number being an integer or operator being out of place). 
# If the input contains an invalid operator output Invalid Operator. If the mathematical operation results in an error output Error.  
# Otherwise output the result of the mathematical operation.

# Sample input: 3 3 +
# Sample output: Incorrect Format
# Sample input: 3 + 3
# Sample output: 6
# Sample input: 3 ? 3
# Sample output: Invalid Operator

def calculate():
    user_input = input()
    parts = user_input.split()

    if len(parts) != 3:
        print("Incorrect Format")
        return

    try:
        num1 = int(parts[0])
        operator = parts[1]
        num2 = int(parts[2])
    except ValueError:
        print("Incorrect Format")
        return

    valid_operators = ['+', '-', '*', '/']
    if operator not in valid_operators:
        print("Invalid Operator")
        return

    try:
        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            if num2 == 0:
                print("Error")  # Division by zero
                return
            result = num1 / num2
        print(int(result)) # Output as integer as per sample
    except Exception:
        print("Error")

calculate()
```

```python
# Define a function called fileInfo that takes one parameter (a string) which is a filename.  
# The function should open and read the file and then display the number of lines, words, and characters in the file. 
# Newline and space characters should not be included in the character count.

# Sample output:
# Lines: 2
# Words: 10
# Characters: 32



def fileInfo(filename):
    with open(filename, 'r') as f:
        lines = 0
        words = 0
        characters = 0

        for line in f:
            lines += 1
            words += len(line.split())
            characters += len("".join(char for char in line if char not in ('\n', ' ')))

        print(f"Lines: {lines}\nWords: {words}\nCharacters: {characters}")
```

```python
# Write a program that takes some input and writes some information to a file.  This is what the program needs to do:
# 1. Ask for user input which is to be used as a filename that you will write to.
# 2. Open the filename from step 1 so that you can write to it as if it does not exist.
# 3. Ask for user input which is one number. This represents the number of remaining lines of input (i.e. if the input is 3, then there will be 3 more lines of input.
# 4. Accept input from the user the number of times indicated from step 3.
# 5. Each input from step 3 is space separated numbers (to be treated as integers).
# 6. Add all integers from each step and write this number to its own line in the file. (i.e. if there are 3 lines of input there should be 3 lines in the file where each line is 1 number.


def process_and_write_to_file():
    filename = input()

    with open(filename, 'w') as file:
          
        num_lines = int(input())

        for i in range(num_lines):
            line_input = input()

            numbers = [int(num) for num in line_input.split()]
                
            line_sum = sum(numbers)
            file.write(str(line_sum) + '\n')

process_and_write_to_file()
```
